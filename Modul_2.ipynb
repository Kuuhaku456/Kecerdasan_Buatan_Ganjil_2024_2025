## DATA VISUALIZATION

```python
df2 = pd.read_csv('Iris.csv')
df2
```

```python
df2['Species'].value_counts()
```

membagi data berdasarkan rata-rata

```python
dataset = df2.groupby('Species').mean()
dataset
```

Menampilkan diagram batang berdasarkan dataset yang sudah di bagi

```python
x = dataset.index
y = dataset.PetalLengthCm
plt.bar(x, y, color=['green', 'blue', 'yellow'])
plt.title('Rata-rata panjang sepal iris')
plt.xlabel('jenis iris')
plt.ylabel('panjang rata-rata sepal')
plt.show()
```

menampilkan grafik histogram

```python
plt.hist(df2['PetalWidthCm'], bins=15, color='skyblue')
plt.title('Histogram panjang petal')
plt.ylabel('Frekuensi kemunculan')
plt.xlabel('Panjang petal')
plt.show()
```

menampilkan data menggunakan pie chart

```python
data = df2.Species.value_counts()
plt.pie(data, labels=data.index, autopct="%1.1f%%")
plt.title('Jumlah Data')
plt.show()
```

menampilkan data menggunakan scatter plot

```python
plt.scatter(df2['SepalLengthCm'], df2['SepalWidthCm'])
plt.title('Ukuran Sepal')
plt.xlabel('Panjang sepal')
plt.ylabel('Lebar sepal')
plt.show()
```

membaca dataset untuk line plot

```python
df3 = pd.read_csv('DailyDelhiClimateTest.csv')
df3.head()
```

```python
df3.info()
```

menampilkan line plot

```python
plt.figure(figsize=(10, 5))
plt.plot(pd.to_datetime(df3.date), df3.humidity)
#              x                       y
plt.xticks(rotation=60, horizontalalignment='right', fontsize=10)
plt.show()
```

```python
cor_matrix = df2.corr(numeric_only = True)
sns.heatmap(cor_matrix, cmap='coolwarm', annot=True, center=0)
plt.show()
```
